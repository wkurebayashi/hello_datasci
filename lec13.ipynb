```
# ライブラリの読み込み
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.read_csv('CO2.csv') # データの読み込み
df.head(10) # 読み込んだデータの先頭10行を表示
```

```
df.plot(x='Date', y='CO2') # データを折れ線グラフとして描画 (横軸が"Date"列、縦軸が"CO2"列)
plt.xticks(rotation=45) # 横軸の目盛の数値を見やすいように傾ける
plt.ylim((0, 450)) # 縦軸の範囲を (0, 450) とする
```

```
df['CO2'] = df['CO2'].rolling(window=12).mean() # 12ヶ月の幅で移動平均
```

```
df.plot(x='Date', y='CO2') # もう一度グラフを描画する
plt.xticks(rotation=45)
plt.ylim((0, 450))
```
